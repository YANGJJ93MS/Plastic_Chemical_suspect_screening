```python
import requests
import pandas as pd

rtfile = pd.read_csv('D:/UCSF_postdoc_topic/REVEAL_topics/REVEAL_200samples_analysis/RT_prediction/RT_library_DTSC_CAS.csv')
cas_numbers = rtfile['CAS'].tolist()

def cas_to_smiles(cas_number):
  
    url = f"https://cactus.nci.nih.gov/chemical/structure/{cas_number}/smiles"
    
    try:
        response = requests.get(url)
        response.raise_for_status()
        smiles = response.text.strip()
        return smiles if smiles else None
      
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data for CAS {cas_number}: {e}")
        return None

if __name__ == "__main__":

    cas_numbers = cas_numbers
    
    print("Converting CAS numbers to SMILES...")
    molecules = [ cas_to_smiles(cas) for cas in cas_numbers ] 
    print("Done.")

    rtfile['SMILES_cas'] = molecules
    print("\nResults:")
```

```python
import tabulate
print(tabulate.tabulate(rtfile, headers='keys', tablefmt='pretty'))
rtfile.to_csv('D:/UCSF_postdoc_topic/REVEAL_topics/REVEAL_200samples_analysis/RT_prediction/RT_library_DTSC_SMILES_fromCAS.csv', index=False)
```
