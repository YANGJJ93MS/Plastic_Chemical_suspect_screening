```python
from rdkit import Chem
from rdkit.Chem import Descriptors
from rdkit.Chem.rdmolops import RemoveHs


# Function to sanitize the SMILES
def sanitize_smiles(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            return None, None

        # Remove fragments after removing salts
        frags = Chem.GetMolFrags(mol, asMols=True, sanitizeFrags=False)
        mol = max(frags, default=None, key=lambda m: m.GetNumAtoms())
        if mol is None:
            return None, None

        # Attempt to sanitize molecule
        Chem.SanitizeMol(mol)

        # Standardize and calculate monoisotopic mass
        standardized_smiles = Chem.MolToSmiles(mol, isomericSmiles=True)
        monoisotopic_mass = Descriptors.ExactMolWt(mol)
        
        return standardized_smiles, monoisotopic_mass
    except Exception as e:
        print(f"Error sanitizing SMILES {smiles}: {e}")
        return None, None

SMILES_ready =[]
Monoisotopic_Mass_ready =[]
for smiles in chemical_df['SMILES']:
    standardized_smiles, monoisotopic_mass = sanitize_smiles(smiles)
    if standardized_smiles:
        SMILES_ready.append(standardized_smiles)
        Monoisotopic_Mass_ready.append(monoisotopic_mass)
    else:
        SMILES_ready.append('NA')
        Monoisotopic_Mass_ready.append('NA')
chemical_df['SMILES_ready'] = SMILES_ready
chemical_df['Monoisotopic_Mass_ready'] = Monoisotopic_Mass_ready
```

```python
#clean up targeted molecule list, retain organic molecules only
import pandas as pd
from rdkit import Chem

# # Retain rows with available value in SMILES column
chemical_df = chemical_df[chemical_df['SMILES_ready'].notna()]

# Convert 'MONOISOTOPIC MASS' column to numeric, forcing errors to NaN
chemical_df['Monoisotopic_Mass_ready'] = pd.to_numeric(chemical_df['Monoisotopic_Mass_ready'], errors='coerce')

# Drop rows with NaN in 'MONOISOTOPIC MASS'
chemical_df = chemical_df[chemical_df['Monoisotopic_Mass_ready'].notna()]

# Function to strictly determine if a molecule is organic
def is_organic(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return False  # invalid SMILES
    organic_elements = {'C', 'H', 'N', 'O', 'P', 'S', 'F', 'Cl', 'Br', 'I'}  # Common organic elements
    inorganic_elements = {'Li', 'Be', 'Na', 'Mg', 'Al', 'K', 'Ca', 'Sc', 'Ti', 'V', 'Cr', 'Mn', 'Fe', 'Co', 'Ni', 'Cu', 'Zn', 'Ga', 'Rb', 'Sr', 'Y', 'Zr', 'Nb', 'Mo', 'Tc', 'Ru', 'Rh', 'Pd', 'Ag', 'Cd', 'In', 'Sn', 'Sb', 'Cs', 'Ba', 'La', 'Ce', 'Pr', 'Nd', 'Pm', 'Sm', 'Eu', 'Gd', 'Tb', 'Dy', 'Ho', 'Er', 'Tm', 'Yb', 'Lu', 'Hf', 'Ta', 'W', 'Re', 'Os', 'Ir', 'Pt', 'Au', 'Hg', 'Tl', 'Pb', 'Bi', 'Th', 'U'}
    contains_carbon = False
    for atom in mol.GetAtoms():
        symbol = atom.GetSymbol()
        if symbol == 'C':
            contains_carbon = True
        if symbol in inorganic_elements:
            return False  # Contains metal or other inorganic elements
        if symbol not in organic_elements:
            return False  # Contains uncommon elements for organic molecules
    return contains_carbon

# Step 1: Filter rows with available MW (MONOISOTOPIC MASS) between 100 and 1000 and classified as organic
filtered_df = []
for index, row in chemical_df.iterrows():
    smiles = row['SMILES_ready']  # Assuming the SMILES column is named 'SMILES'
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        mw = row['Monoisotopic_Mass_ready']  # Assuming the column is named 'MONOISOTOPIC MASS'
        if 100 <= mw <= 1000:
            if is_organic(smiles):
                filtered_df.append(row)

# Create a new DataFrame from the filtered rows
filtered_df = pd.DataFrame(filtered_df)

# Save the filtered DataFrame to a new CSV file
# filtered_df.to_csv('D:/UCSF_postdoc_topic/ECHO_project/Reprocessed_MSDIAL/Filtered_Chemical_List.csv', index=False)

print("Filtering complete. The filtered list has been saved as 'Filtered_Chemical_List.csv'.")
```

```python
import pandas as pd
import numpy as np

def process_msdial_file(file_path):
    """
    Processes an MSDIAL output file to extract sample metadata and peak area data.
    
    Parameters:
        file_path (str): Path to the MSDIAL output file.
        
    Returns:
        tuple: Two DataFrames - (sample_meta, peak_area_data)
            - sample_meta: Metadata about the samples (Class, File type, Injection order, Batch ID).
            - peak_area_data: Data on peak areas with columns "Average" and "Stdev" retained at the end.
    """
    
    # Step 1: Load the data with all columns as strings to avoid parsing issues
    data = pd.read_csv(file_path, delimiter='\t', header=None, dtype=str, keep_default_na=False)
    
    # Step 2: Extract sample metadata
    sample_meta_raw = data.iloc[0:5, :]  # Selecting the first four rows for sample metadata

    # Transpose and clean the metadata
    sample_meta = sample_meta_raw.transpose()  # Transpose to switch rows to columns
    sample_meta.columns = ["Class", "File type", "Injection order", "Batch ID", "Sample_ID"]  # Rename columns

    # Remove rows with all NA or irrelevant values such as "NA" or ""
    sample_meta = sample_meta.dropna(how='all')  # Remove rows where all values are NaN or empty
    sample_meta = sample_meta[~sample_meta.isin(["NA", "", "Average", "Stdev"]).any(axis=1)] # Filter out rows with non-informative values
    sample_meta = sample_meta.iloc[1:,:]

    # Step 3: Extract peak area data
    # Starting from row 5 (index 4), we extract the peak area data
    peak_area_data = data.iloc[4:, :].reset_index(drop=True)  # Removing the first four rows for peak area information

    # Set column names from the fifth row (header row) and then remove that row from the data
    peak_area_data.columns = data.iloc[4].tolist()  # Set column names from the fifth row
    peak_area_data = peak_area_data.iloc[1:, :]  # Remove the row used for column names

    # Clean up potential duplicate column names by appending a suffix
    peak_area_data.columns = pd.Series(peak_area_data.columns).apply(
        lambda x: x if pd.Series(peak_area_data.columns).value_counts()[x] == 1 
        else f"{x}_{pd.Series(peak_area_data.columns).value_counts()[x]}"
    ).tolist()

    # Return both DataFrames
    return sample_meta, peak_area_data

# Example usage
file_path = "D:/UCSF_postdoc_topic/REVEAL_topics/REVEAL_200samples_analysis/Neg_AlignmentResults/Area_1_2024_11_03_11_28_00.txt"
file_path2 = "D:/UCSF_postdoc_topic/REVEAL_topics/REVEAL_200samples_analysis/Pos_AlignmentResults/Area_1_2024_10_28_20_25_25.txt"
sample_meta1, peak_area_data1 = process_msdial_file(file_path)
sample_meta2, peak_area_data2 = process_msdial_file(file_path2)

# Display results (optional)
print("Sample Metadata:")
print(sample_meta1)
print("\nPeak Area Data:")
print(peak_area_data1)
```

```python
import pandas as pd
from tqdm import tqdm

def find_chemical_hits(msdial_file, meta_file, filtered_df, ppm_threshold=20):
    """
    Matches compounds in the filtered DataFrame with LCMS processed data based on monoisotopic mass
    within a given ppm threshold. Returns the number of chemicals with hits and a DataFrame of matched chemicals.
    
    Parameters:
        msdial_path (str): Path to the MSDIAL result file with peak areas.
        meta_path (str): Path to the metadata file for filtering QC and Blank samples.
        filtered_df (pd.DataFrame): DataFrame containing chemicals to be matched (with columns including 'MONOISOTOPIC_MASS').
        ppm_threshold (float): Threshold for parts-per-million (ppm) difference in mass matching.
        
    Returns:
        int: Number of chemicals in the filtered DataFrame with hits from the MSDIAL result file.
        pd.DataFrame: DataFrame with detailed matched chemicals and corresponding MSDIAL result rows.
    """

    # Load LCMS processed data with peak areas (MSDIAL result file)
    # msdial_df = pd.read_csv(msdial_path, encoding='ISO-8859-1')
    
    # Retain QC and Blank samples only from metadata
    # metafile = pd.read_csv(meta_path, encoding='ISO-8859-1')
    msdial_df = msdial_file
    metafile = meta_file
    qc_blank_samples = metafile[metafile['File type'].isin(['QC'])]['Sample_ID'].tolist()
    
    # Filter MSDIAL result file for QC and Blank samples based on 'MS/MS spectrum' columns
    qc_blank_msdial_df = msdial_df[['Average Rt(min)', 'Average Mz'] + [col for col in msdial_df.columns if col in qc_blank_samples]]

     # Convert 'Average Mz' column to numeric, handling non-numeric values if any
    qc_blank_msdial_df['Average Mz'] = pd.to_numeric(qc_blank_msdial_df['Average Mz'], errors='coerce')
    filtered_df['MONOISOTOPIC_MASS'] = pd.to_numeric(filtered_df['MONOISOTOPIC_MASS'], errors='coerce')
   
    
    # Function to calculate ppm difference
    def ppm_difference(value1, value2):
        return abs(value1 - value2) / value1 * 1e6

    # Prepare to store matched rows
    matched_rows = []
    
    # Iterate over each compound in filtered_df to find hits within ppm threshold
    filtered_df['Hits'] = 0  # Initialize a 'Hits' column
    for _, chem_row in tqdm(filtered_df.iterrows(), total=filtered_df.shape[0], desc="Matching compounds"):
        mw = chem_row['MONOISOTOPIC_MASS']
        matches = qc_blank_msdial_df[
            qc_blank_msdial_df['Average Mz'].apply(lambda mz: ppm_difference(mz, mw) <= ppm_threshold)
        ]
        if not matches.empty:
            # Update 'Hits' column in filtered_df for matched compounds
            filtered_df.at[chem_row.name, 'Hits'] = len(matches)
            # Append each match with additional compound information
            for _, match_row in matches.iterrows():
                matched_rows.append({
                    **match_row,
                    'Matched Compound': chem_row['SMILES'],
                    'PREFERRED NAME': chem_row.get('PREFERRED NAME', None),
                    'DTXSID': chem_row.get('DTXSID', None)
                })
    
    # Filter filtered_df to only include chemicals with hits
    filtered_with_hits_df = filtered_df[filtered_df['Hits'] > 0]
    
    # Create a new DataFrame for matched MSDIAL result rows
    matched_msdial_df = pd.DataFrame(matched_rows)
    
    # Calculate the number of chemicals with hits
    num_chemicals_with_hits = filtered_with_hits_df.shape[0]
    
    # Return the count and the DataFrame of matched rows
    return num_chemicals_with_hits, matched_msdial_df

# Example usage
# path4 = 'path_to_msdial_result_file.csv'  # Replace with actual MSDIAL file path
# path5 = 'path_to_metadata_file.csv'       # Replace with actual metadata file path

filtered_df = pd.read_csv('D:/UCSF_postdoc_topic/REVEAL_topics/First100_batch/Plastic_Chemical_List_organic.csv')
num_hits2, chemical_hits_df2 = find_chemical_hits(peak_area_data2, sample_meta2, filtered_df, ppm_threshold=20)
num_hits1, chemical_hits_df1 = find_chemical_hits(peak_area_data1, sample_meta1, filtered_df, ppm_threshold=20)

# Display results
print(f"Number of chemicals with hits1: {num_hits1}")
print(f"Number of chemicals with hits2: {num_hits2}")
```
